```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Used for helper functions
import imperial_model
```

```python
predictions = pd.read_csv("submissions/final_submission.csv")
```

```python
dates_of_interest = ['2020-05-25', '2020-05-26', '2020-05-27', '2020-05-28', '2020-05-29', '2020-05-30', '2020-05-31', '2020-06-01', '2020-06-02', '2020-06-03', '2020-06-04', '2020-06-05', '2020-06-06', '2020-06-07']

def get_preds_for_fips(FIPS):
    preds = []
    
    for date in dates_of_interest:
        query = "id==" + "\"" + date + "-" + str(FIPS) + "\""
        median_pred  = list(predictions.query(query).values[0])[5]
        
        preds.append(median_pred)
        
    return np.array(preds)

# Use this function to print out the top 'count' fips by cumulative deaths
def get_large_fips(count):
    return imperial_model.get_top_cumuluative_death_counties(count, "2020-05-24")
```

```python
print(get_large_fips(200))
```

```python
for FIPS in get_large_fips(300):
    actual = imperial_model.imperial_util.get_deaths_list(FIPS, "2020-05-24")[-20:]
    x_actual = [i for i in range(len(actual))]

    preds = get_preds_for_fips(FIPS)
    x_pred = [i + len(actual) for i in range(len(preds))]

    plt.plot(x_actual, actual, color="blue")
    plt.plot(x_pred, preds, color="green")
    plt.show()
```
